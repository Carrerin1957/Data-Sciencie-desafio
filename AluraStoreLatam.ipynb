# 3. Calificación promedio de la tienda


```python
# Promedio por categoría con 2 decimales
promedio_por_producto = tienda.groupby('Categoría del Producto')['Calificación'].mean().reset_index()
promedio_por_producto['Calificación'] = promedio_por_producto['Calificación'].round(2)
print(promedio_por_producto)

# Promedio general de la tienda con 2 decimales
promedio_tienda = tienda['Calificación'].mean()
print(f"El promedio general de la tienda es: {promedio_tienda:.2f}")
```

```python
# Promedio por categoría con 2 decimales
promedio_por_producto2 = tienda2.groupby('Categoría del Producto')['Calificación'].mean().reset_index()
promedio_por_producto2['Calificación'] = promedio_por_producto2['Calificación'].round(2)
print(promedio_por_producto2)

# Promedio general de la tienda con 2 decimales
promedio_tienda2 = tienda2['Calificación'].mean()
print(f"El promedio general de la tienda es: {promedio_tienda2:.2f}")
```

```python
# Promedio por categoría con 2 decimales
promedio_por_producto3 = tienda3.groupby('Categoría del Producto')['Calificación'].mean().reset_index()
promedio_por_producto3['Calificación'] = promedio_por_producto3['Calificación'].round(2)
print(promedio_por_producto3)

# Promedio general de la tienda con 2 decimales
promedio_tienda3 = tienda3['Calificación'].mean()
print(f"El promedio general de la tienda es: {promedio_tienda3:.2f}")
```

# 6. Creación de graficas

```python
import matplotlib.pyplot as plt

# Nombres de las tiendas
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
facturaciones = [total_precio, total_precio2, total_precio3, total_precio4]
```

```python
plt.figure(figsize=(8, 5))
plt.bar(nombres_tiendas, facturaciones, color='orange', edgecolor='black')

plt.title('Facturación total por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Total facturado ($)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

# Mostrar valores encima de cada barra
for i, valor in enumerate(facturaciones):
    plt.text(i, valor + max(facturaciones)*0.01, f"${valor:,.2f}", ha='center', va='bottom')

plt.show()
```

```python

# Asumiendo que ya tienes esto:
# conteo_categorias = tienda['Categoría del Producto'].value_counts()

# Crear gráfico de pastel
plt.figure(figsize=(8, 8))  # Tamaño opcional
plt.pie(
    conteo_categorias,
    labels=conteo_categorias.index,
    autopct='%1.1f%%',       # Muestra el porcentaje con 1 decimal
    startangle=90,           # Comienza desde el ángulo superior
    colors=plt.cm.Set3.colors,  # Paleta de colores (opcional)
    wedgeprops={'edgecolor': 'black'}  # Borde negro a cada porción
)

plt.title('Distribución de productos por categoría')
plt.axis('equal')  # Asegura que el pastel sea circular

plt.show()
```

```python

# Datos
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costos_envio = [costo_envio, costo_envio2, costo_envio3, costo_envio4]

# Crear gráfico de barras horizontales
plt.figure(figsize=(10, 5))
plt.barh(nombres_tiendas, costos_envio, color='mediumseagreen', edgecolor='black')

# Etiquetas y título
plt.title('Gastos de envío por tienda')
plt.xlabel('Costo total ($)')
plt.ylabel('Tiendas')
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Mostrar valores al lado de cada barra
for i, valor in enumerate(costos_envio):
    plt.text(valor + max(costos_envio)*0.01, i, f"${valor:,.2f}", va='center')

plt.tight_layout()
plt.show()
```
